```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
from sklearn import tree
```

```python
clf=tree.DecisionTreeClassifier(random_state=0)
iris=load_iris()
clf=clf.fit(iris.data,iris.target)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(clf,filled=True)

#tree.plot_tree is a new fucntion introduced. Previously we used to use graphviz
```

```python
print(tree.export_text(clf))
```

```python
#using random forest
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier(n_estimators=20,random_state=0)
iris=load_iris()
clf=clf.fit(iris.data,iris.target)
#n_estimator by deault takes as 100 if not mentioned.
```

```python
#or we can do like the old way where we use x with independent feature and y with dependent as shown below
clf=RandomForestClassifier(n_estimators=20,random_state=0)
iris=load_iris()
X=iris.data
y=iris.target

clf=clf.fit(X,y)
```

```python
#we can get all the information on the decision tree from the below fucntion
clf.estimators_
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(clf.estimators_[2],filled=True)

#in clf_estimators we can change the index to view the required tree
#random forest uses bootstrap aggregation so it will randomly select the features.
```

```python
plt.figure(figsize=(15,10))
for i in range(len(clf.estimators_)):
     tree.plot_tree(clf.estimators_[i],filled=True)
```
